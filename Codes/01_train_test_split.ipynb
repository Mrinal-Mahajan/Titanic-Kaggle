```python
import pandas as pd
import os
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
os.chdir('../Input_Data')
```

```python
df = pd.read_csv('train.csv',header=0)
```

```python
y = np.array(df['SalePrice'])
```

```python
x = np.array(df.drop(['SalePrice'],axis=1))
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
df_train = pd.DataFrame(x_train,columns=df.drop(['SalePrice'],axis=1).columns)
```

```python
df_test = pd.DataFrame(x_test,columns=df.drop(['SalePrice'],axis=1).columns)
```

```python
df_train['SalePrice'] = y_train
```

```python
df_test['SalePrice'] = y_test
```

```python
os.chdir('../output_data')
```

```python
df_train.to_csv('train.csv',index=False)
```

```python
df_test.to_csv('validation.csv',index=False)
```
